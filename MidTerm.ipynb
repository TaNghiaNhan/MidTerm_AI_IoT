```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.autograd import Variable
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import glob
from scipy import signal
import re

# Đường dẫn đến thư mục dữ liệu BIDMC
BIDMC_PATH = '/content/drive/MyDrive/codePPG2025/bidmc/bidmc_csv'
# Đường dẫn lưu dữ liệu đã xử lý
PROCESSED_PATH = '/content/drive/MyDrive/codePPG2025/bidmc/processed'
# Đường dẫn lưu mô hình
MODEL_SAVE_PATH = '/content/drive/MyDrive/codePPG2025/bidmc/models'
# Đường dẫn lưu kết quả
RESULTS_PATH = '/content/drive/MyDrive/codePPG2025/bidmc/results'

# Tạo thư mục nếu chưa tồn tại
for path in [PROCESSED_PATH, MODEL_SAVE_PATH, RESULTS_PATH]:
    os.makedirs(path, exist_ok=True)

# Thiết lập seed để kết quả có thể tái tạo
torch.manual_seed(42)
np.random.seed(42)

# Kiểm tra và sử dụng GPU nếu có
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Sử dụng thiết bị: {device}")

# ===================== PHẦN XỬ LÝ DỮ LIỆU =====================

def list_bidmc_files():
    """
    Liệt kê các file trong bộ dữ liệu BIDMC và nhóm chúng theo ID bệnh nhân
    """
    all_files = os.listdir(BIDMC_PATH)

    # Tạo từ điển nhóm file theo ID bệnh nhân
    patient_files = {}

    for file in all_files:
        # Tìm ID bệnh nhân từ tên file
        match = re.search(r'bidmc_(\d+)', file)
        if match:
            patient_id = str(int(match.group(1)))  # Loại bỏ số 0 ở đầu
            if patient_id not in patient_files:
                patient_files[patient_id] = {'signals': None, 'numerics': None, 'breaths': None}

            if 'Signal' in file:
                patient_files[patient_id]['signals'] = file
            elif 'Numerics' in file:
                patient_files[patient_id]['numerics'] = file
            elif 'Breaths' in file:
                patient_files[patient_id]['breaths'] = file
    return patient_files

def load_patient_data(patient_id, patient_files):
    """
    Tải dữ liệu cho một bệnh nhân cụ thể
    """

    files = patient_files[patient_id]
    data = {}

    # Tải file tín hiệu
    if files['signals']:
        signals_path = os.path.join(BIDMC_PATH, files['signals'])
        data['signals'] = pd.read_csv(signals_path)

    # Tải file thông số
    if files['numerics']:
        numerics_path = os.path.join(BIDMC_PATH, files['numerics'])
        data['numerics'] = pd.read_csv(numerics_path)

    # Tải file nhịp thở
    if files['breaths']:
        breaths_path = os.path.join(BIDMC_PATH, files['breaths'])
        data['breaths'] = pd.read_csv(breaths_path)

    return data

def apply_bandpass_filter(data, fs=125, lowcut=0.5, highcut=8.0, order=4):
    """
    Áp dụng bộ lọc Butterworth băng thông cho tín hiệu
    """
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = signal.butter(order, [low, high], btype='band')
    return signal.filtfilt(b, a, data)

def normalize_signal(data):
    """
    Chuẩn hóa tín hiệu về khoảng 0-1
    """
    min_val = np.min(data)
    max_val = np.max(data)

    # Kiểm tra để tránh chia cho 0
    if max_val == min_val:
        return np.zeros_like(data)

    return (data - min_val) / (max_val - min_val)

def segment_data(data, fs=125, window_size=10):
    """
    Phân đoạn dữ liệu thành các cửa sổ có kích thước cố định
    """
    # Số mẫu trong mỗi cửa sổ
    samples_per_window = int(fs * window_size)

    # Số cửa sổ hoàn chỉnh có thể tạo ra
    num_windows = len(data) // samples_per_window

    # Phân đoạn dữ liệu
    segments = []
    for i in range(num_windows):
        start_idx = i * samples_per_window
        end_idx = start_idx + samples_per_window
        segment = data[start_idx:end_idx]
        segments.append(segment)

    return segments, [i * window_size for i in range(num_windows)]

# def match_numerics_to_segment(segment_times, numerics_df):
#     """
#     Lấy giá trị HR và RR từ bảng numerics cho mỗi phân đoạn
#     """
#     hr_values = []
#     rr_values = []

#     for start_time in segment_times:
#         # Tìm các giá trị trong khoảng thời gian của phân đoạn
#         segment_numerics = numerics_df[(numerics_df['Time [s]'] >= start_time) &
#                                        (numerics_df['Time [s]'] < start_time + 10)]

#         if len(segment_numerics) > 0:
#             # Lấy giá trị trung bình trong đoạn
#             hr = segment_numerics[' HR'].mean()
#             rr = segment_numerics[' RESP'].mean()
#         else:
#             # Nếu không có giá trị nào, lấy giá trị gần nhất
#             closest_idx = (numerics_df['Time [s]'] - start_time).abs().idxmin()
#             hr = numerics_df.loc[closest_idx, ' HR']
#             rr = numerics_df.loc[closest_idx, ' RESP']
#         hr_values.append(hr)
#         rr_values.append(rr)

#     return np.array(hr_values), np.array(rr_values)

def match_numerics_to_segment(segment_times, numerics_df):
    """
    Lấy giá trị HR và RR từ bảng numerics cho mỗi phân đoạn
    """
    # Điền giá trị trung bình nếu cột RR bị thiếu
    if ' RESP' not in numerics_df.columns:
        mean_rr = 12.0  # Giá trị trung bình mặc định
        numerics_df[' RESP'] = mean_rr

    hr_values = []
    rr_values = []

    for start_time in segment_times:
        # Tìm các giá trị trong khoảng thời gian của phân đoạn
        segment_numerics = numerics_df[(numerics_df['Time [s]'] >= start_time) &
                                       (numerics_df['Time [s]'] < start_time + 10)]

        if len(segment_numerics) > 0:
            # Lấy giá trị trung bình trong đoạn
            hr = segment_numerics[' HR'].mean()
            rr = segment_numerics[' RESP'].mean()
        else:
            # Nếu không có giá trị nào, lấy giá trị gần nhất
            closest_idx = (numerics_df['Time [s]'] - start_time).abs().idxmin()
            hr = numerics_df.loc[closest_idx, ' HR']
            rr = numerics_df.loc[closest_idx, ' RESP']

        hr_values.append(hr)
        rr_values.append(rr)

    return np.array(hr_values), np.array(rr_values)

def process_bidmc_data():
    """
    Xử lý tất cả dữ liệu BIDMC và lưu các phân đoạn đã xử lý
    """
    # Liệt kê tất cả các file
    patient_files = list_bidmc_files()

    all_segments = {
        'ppg': [],  # PLETH signal
        'ecg': [],  # Lead II (ECG)
        'resp': [], # RESP signal
        'hr': [],   # Heart rate
        'rr': []    # Respiratory rate
    }

    def fill_missing_rr(numerics_df):
        # Kiểm tra và điền giá trị trung bình nếu RR bị thiếu
        if ' RESP' not in numerics_df.columns:
            print("Cảnh báo: Cột RESP không tồn tại. Tạo cột mới với giá trị trung bình.")
            # Tìm giá trị trung bình của RR từ toàn bộ dữ liệu
            mean_rr = 12.0  # Giá trị trung bình mặc định nếu không có dữ liệu
            numerics_df[' RESP'] = mean_rr
        else:
            # Điền giá trị trung bình cho các ô trống
            mean_rr = numerics_df[' RESP'].mean()
            numerics_df[' RESP'].fillna(mean_rr, inplace=True)

        return numerics_df

    for patient_id, files in patient_files.items():
        print(f"Đang xử lý dữ liệu bệnh nhân {patient_id}...")

        # Kiểm tra xem có đủ file không
        if not (files['signals'] and files['numerics']):
            print(f"Thiếu file cho bệnh nhân {patient_id}, bỏ qua.")
            continue

        # Tải dữ liệu
        data = load_patient_data(patient_id, patient_files)

        # Xử lý tín hiệu
        # Lấy các cột tín hiệu
        ppg = data['signals'][' PLETH'].values  # Tín hiệu PPG
        ecg = data['signals'][' II'].values     # Tín hiệu ECG (lead II)
        resp = data['signals'][' RESP'].values  # Tín hiệu hô hấp

        # Lọc băng thông
        ppg_filtered = apply_bandpass_filter(ppg)
        ecg_filtered = apply_bandpass_filter(ecg)
        resp_filtered = apply_bandpass_filter(resp, lowcut=0.05, highcut=1.0)  # Lọc khác cho hô hấp

        # Chuẩn hóa
        ppg_normalized = normalize_signal(ppg_filtered)
        ecg_normalized = normalize_signal(ecg_filtered)
        resp_normalized = normalize_signal(resp_filtered)

        # Phân đoạn
        ppg_segments, segment_times = segment_data(ppg_normalized)
        ecg_segments, _ = segment_data(ecg_normalized)
        resp_segments, _ = segment_data(resp_normalized)

        data['numerics'] = fill_missing_rr(data['numerics'])  # Hàm fill_missing_rr được định nghĩa như đã đề cập ở trên

        # Lấy thông số HR và RR cho mỗi phân đoạn
        hr_values, rr_values = match_numerics_to_segment(segment_times, data['numerics'])

        # Thêm vào danh sách tổng
        all_segments['ppg'].extend(ppg_segments)
        all_segments['ecg'].extend(ecg_segments)
        all_segments['resp'].extend(resp_segments)
        all_segments['hr'].extend(hr_values)
        all_segments['rr'].extend(rr_values)

        print(f"Đã xử lý {len(ppg_segments)} phân đoạn cho bệnh nhân {patient_id}.")

    # Chuyển đổi thành mảng numpy
    for key in all_segments:
        all_segments[key] = np.array(all_segments[key])

    # Lưu dữ liệu đã xử lý
    np.save(os.path.join(PROCESSED_PATH, 'bidmc_processed_segments.npy'), all_segments)

    print(f"Đã hoàn thành xử lý {len(all_segments['ppg'])} phân đoạn từ tất cả bệnh nhân.")

    # Vẽ một vài ví dụ để kiểm tra
    visualize_processed_data(all_segments)

    return all_segments

def visualize_processed_data(segments, num_examples=3):
    """
    Trực quan hóa một số phân đoạn đã xử lý
    """
    # Chọn ngẫu nhiên một số phân đoạn
    indices = np.random.choice(len(segments['ppg']), num_examples, replace=False)

    fig, axs = plt.subplots(num_examples, 3, figsize=(15, 5 * num_examples))

    for i, idx in enumerate(indices):
        # Vẽ PPG
        axs[i, 0].plot(segments['ppg'][idx])
        axs[i, 0].set_title(f'PPG - Phân đoạn {idx}')

        # Vẽ ECG
        axs[i, 1].plot(segments['ecg'][idx])
        axs[i, 1].set_title(f'ECG - Phân đoạn {idx}')

        # Vẽ RESP
        axs[i, 2].plot(segments['resp'][idx])
        axs[i, 2].set_title(f'RESP - Phân đoạn {idx}')

        # Thêm thông tin HR và RR
        axs[i, 0].set_xlabel(f'HR: {segments["hr"][idx]:.1f} BPM, RR: {segments["rr"][idx]:.1f} BrPM')

    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_PATH, 'processed_data_examples.png'))
    plt.close()

# ===================== PHẦN ĐỊNH NGHĨA DATASET =====================

class BIDMCDatasetWithHRBR(Dataset):
    """
    Dataset cho dữ liệu BIDMC với thông số HR và BR
    """
    def __init__(self, data_path=PROCESSED_PATH, filename='bidmc_processed_segments.npy'):
        """
        Khởi tạo dataset
        """
        # Tải dữ liệu đã xử lý
        file_path = os.path.join(data_path, filename)
        self.segments = np.load(file_path, allow_pickle=True).item()

        # Handle NaN values in RR
        rr_data = self.segments['rr']
        # Replace NaN with the mean of valid RR values. This approach maintains the validity of other samples while avoiding errors due to NaN values.
        # Change fill_value according to your dataset and how you want to deal with NaN values.
        rr_data = np.nan_to_num(rr_data, nan=np.nanmean(rr_data))
        self.segments['rr'] = rr_data  #Update the segment data with the modified rr_data.

        self.hr_mean, self.hr_std = np.mean(self.segments['hr']), np.std(self.segments['hr'])
        self.rr_mean, self.rr_std = np.mean(self.segments['rr']), np.std(self.segments['rr'])
        # Lấy số lượng phân đoạn
        self.num_segments = len(self.segments['ppg'])

        # Chuẩn hóa HR và BR
        print("rr : ",self.segments['rr'])
        self.hr_mean, self.hr_std = np.mean(self.segments['hr']), np.std(self.segments['hr'])
        self.rr_mean, self.rr_std = np.mean(self.segments['rr']), np.std(self.segments['rr'])

        self.normalized_hr = (self.segments['hr'] - self.hr_mean) / (self.hr_std + 1e-6)
        self.normalized_rr = (self.segments['rr'] - self.rr_mean) / (self.rr_std + 1e-6)

        print(f"Dataset được khởi tạo với {self.num_segments} phân đoạn")
        print(f"Phạm vi HR: {np.min(self.segments['hr']):.1f} - {np.max(self.segments['hr']):.1f} BPM, Trung bình: {self.hr_mean:.1f}, Độ lệch chuẩn: {self.hr_std:.1f}")
        print(f"Phạm vi RR: {np.min(self.segments['rr']):.1f} - {np.max(self.segments['rr']):.1f} BrPM, Trung bình: {self.rr_mean:.1f}, Độ lệch chuẩn: {self.rr_std:.1f}")

    def __len__(self):
        return self.num_segments

    def __getitem__(self, idx):
        # Lấy các phân đoạn
        ppg = self.segments['ppg'][idx]
        ecg = self.segments['ecg'][idx]
        resp = self.segments['resp'][idx]
        hr = self.normalized_hr[idx]
        rr = self.normalized_rr[idx]

        # Chuyển thành tensor
        ppg_tensor = torch.FloatTensor(ppg)
        ecg_tensor = torch.FloatTensor(ecg)
        resp_tensor = torch.FloatTensor(resp)
        hr_tensor = torch.FloatTensor([hr])
        rr_tensor = torch.FloatTensor([rr])

        return ecg_tensor, ppg_tensor, resp_tensor, hr_tensor, rr_tensor

    def denormalize_params(self, hr, rr):
        """
        Chuyển đổi HR và RR từ dạng chuẩn hóa về dạng gốc
        """
        hr_denorm = hr * self.hr_std + self.hr_mean
        rr_denorm = rr * self.rr_std + self.rr_mean
        return hr_denorm, rr_denorm

def prepare_dataloaders(batch_size=32, test_size=0.2):
    """
    Chuẩn bị DataLoader cho tập huấn luyện và kiểm thử
    """
    # Kiểm tra xem file dữ liệu đã xử lý có tồn tại không
    file_path = os.path.join(PROCESSED_PATH, 'bidmc_processed_segments.npy')
    if not os.path.exists(file_path):
        print("Dữ liệu đã xử lý không tồn tại. Đang xử lý dữ liệu...")
        process_bidmc_data()

    # Tạo dataset
    dataset = BIDMCDatasetWithHRBR()

    # Chia tập huấn luyện và kiểm thử
    train_indices, val_indices = train_test_split(
        range(len(dataset)),
        test_size=test_size,
        random_state=42
    )

    # Tạo DataLoader
    train_loader = DataLoader(
        dataset,
        batch_size=batch_size,
        sampler=torch.utils.data.SubsetRandomSampler(train_indices),
        num_workers=2,
        pin_memory=True
    )

    val_loader = DataLoader(
        dataset,
        batch_size=batch_size,
        sampler=torch.utils.data.SubsetRandomSampler(val_indices),
        num_workers=2,
        pin_memory=True
    )

    return train_loader, val_loader, dataset

# ===================== PHẦN ĐỊNH NGHĨA MÔ HÌNH =====================

class EncoderWithParams(nn.Module):
    """
    Encoder cho mô hình cVAE với thông số HR và BR
    """
    def __init__(self, input_dim=1250, hidden_dim=512, latent_dim=128, condition_dim=1250):
        super(EncoderWithParams, self).__init__()

        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.latent_dim = latent_dim

        # Mạng Neural cho input signal (PPG)
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.bn1 = nn.BatchNorm1d(hidden_dim)

        # Mạng Neural cho conditional signal (ECG)
        self.cond_fc = nn.Linear(condition_dim, hidden_dim)
        self.cond_bn = nn.BatchNorm1d(hidden_dim)

        # Mạng Neural cho tín hiệu hô hấp
        self.resp_fc = nn.Linear(condition_dim, hidden_dim // 2)
        self.resp_bn = nn.BatchNorm1d(hidden_dim // 2)

        # Mạng Neural cho thông số HR và BR
        self.param_fc = nn.Linear(2, hidden_dim // 4)
        self.param_bn = nn.BatchNorm1d(hidden_dim // 4)

        # Mạng kết hợp
        combined_dim = hidden_dim * 2 + hidden_dim // 2 + hidden_dim // 4
        self.fc2 = nn.Linear(combined_dim, hidden_dim)
        self.bn2 = nn.BatchNorm1d(hidden_dim)

        # Lớp cho mean và variance
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)
        self.fc_var = nn.Linear(hidden_dim, latent_dim)

    def forward(self, x, c, r, hr, br):
        # Kết hợp HR và BR
        params = torch.cat([hr, br], dim=1)

        # Xử lý input signal (PPG)
        x = F.relu(self.bn1(self.fc1(x)))

        # Xử lý conditional signal (ECG)
        c = F.relu(self.cond_bn(self.cond_fc(c)))

        # Xử lý tín hiệu hô hấp
        r = F.relu(self.resp_bn(self.resp_fc(r)))

        # Xử lý thông số
        p = F.relu(self.param_bn(self.param_fc(params)))

        # Kết hợp đặc trưng
        h = torch.cat([x, c, r, p], dim=1)
        h = F.relu(self.bn2(self.fc2(h)))

        # Tính mean và variance
        mu = self.fc_mu(h)
        log_var = self.fc_var(h)

        return mu, log_var

class DecoderWithParams(nn.Module):
    """
    Decoder cho mô hình cVAE với thông số HR và BR
    """
    def __init__(self, latent_dim=128, hidden_dim=512, output_dim=1250, condition_dim=1250):
        super(DecoderWithParams, self).__init__()

        self.latent_dim = latent_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim

        # Mạng Neural cho latent vector
        self.fc1 = nn.Linear(latent_dim, hidden_dim)
        self.bn1 = nn.BatchNorm1d(hidden_dim)

        # Mạng Neural cho conditional signal (ECG)
        self.cond_fc = nn.Linear(condition_dim, hidden_dim)
        self.cond_bn = nn.BatchNorm1d(hidden_dim)

        # Mạng Neural cho tín hiệu hô hấp
        self.resp_fc = nn.Linear(condition_dim, hidden_dim // 2)
        self.resp_bn = nn.BatchNorm1d(hidden_dim // 2)

        # Mạng Neural cho thông số HR và BR
        self.param_fc = nn.Linear(2, hidden_dim // 4)
        self.param_bn = nn.BatchNorm1d(hidden_dim // 4)

        # Mạng kết hợp
        combined_dim = hidden_dim * 2 + hidden_dim // 2 + hidden_dim // 4
        self.fc2 = nn.Linear(combined_dim, hidden_dim)
        self.bn2 = nn.BatchNorm1d(hidden_dim)

        # Lớp đầu ra
        self.fc3 = nn.Linear(hidden_dim, output_dim)

    def forward(self, z, c, r, hr, br):
        # Kết hợp HR và BR
        params = torch.cat([hr, br], dim=1)

        # Xử lý latent vector
        z = F.relu(self.bn1(self.fc1(z)))

        # Xử lý conditional signal (ECG)
        c = F.relu(self.cond_bn(self.cond_fc(c)))

        # Xử lý tín hiệu hô hấp
        r = F.relu(self.resp_bn(self.resp_fc(r)))

        # Xử lý thông số
        p = F.relu(self.param_bn(self.param_fc(params)))

        # Kết hợp đặc trưng
        h = torch.cat([z, c, r, p], dim=1)
        h = F.relu(self.bn2(self.fc2(h)))

        # Đầu ra
        output = torch.sigmoid(self.fc3(h))  # Sigmoid do dữ liệu đã được chuẩn hóa về 0-1

        return output

class ConditionalVAEWithParams(nn.Module):
    """
    Mô hình Conditional Variational Autoencoder (cVAE) với thông số HR và BR
    """
    def __init__(self, input_dim=1250, hidden_dim=512, latent_dim=128, condition_dim=1250):
        super(ConditionalVAEWithParams, self).__init__()

        self.latent_dim = latent_dim

        # Encoder và Decoder
        self.encoder = EncoderWithParams(input_dim, hidden_dim, latent_dim, condition_dim)
        self.decoder = DecoderWithParams(latent_dim, hidden_dim, input_dim, condition_dim)

    def reparameterize(self, mu, log_var):
        """
        Reparameterization trick để cho phép backpropagation qua quá trình lấy mẫu
        """
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        z = mu + eps * std
        return z

    def forward(self, x, c, r, hr, br):
        # Encode
        mu, log_var = self.encoder(x, c, r, hr, br)

        # Reparameterize
        z = self.reparameterize(mu, log_var)

        # Decode
        recon_x = self.decoder(z, c, r, hr, br)

        return recon_x, mu, log_var

    def generate(self, c, r, hr, br, num_samples=1):
        """
        Sinh tín hiệu từ điều kiện và thông số

        Tham số:
            c: Tín hiệu ECG
            r: Tín hiệu hô hấp
            hr: Nhịp tim chuẩn hóa
            br: Nhịp thở chuẩn hóa
            num_samples: Số lượng mẫu cần sinh
        """
        # Chuyển về thiết bị chính xác
        c = c.to(device)
        r = r.to(device)
        hr = hr.to(device)
        br = br.to(device)

        # Tạo latent vectors ngẫu nhiên
        z = torch.randn(num_samples, self.latent_dim).to(device)

        # Decode để tạo tín hiệu
        with torch.no_grad():
            # Đảm bảo hr và br có số chiều phù hợp
            if len(hr.shape) == 1:
                hr = hr.unsqueeze(1)
            if len(br.shape) == 1:
                br = br.unsqueeze(1)

            generated = self.decoder(z, c, r, hr, br)

        return generated

def vae_loss_function(recon_x, x, mu, log_var, beta=1.0):
    """
    Hàm mất mát cho VAE

    Tham số:
        recon_x: Tín hiệu tái tạo
        x: Tín hiệu gốc
        mu: Mean của phân phối latent
        log_var: Log variance của phân phối latent
        beta: Trọng số cho KL divergence

    Trả về:
        loss: Tổng mất mát
        recon_loss: Mất mát tái tạo
        kl_loss: KL divergence
    """
    # Reconstruction loss (MSE)
    recon_loss = F.mse_loss(recon_x, x, reduction='sum')

    # KL divergence
    kl_loss = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())

    # Tổng mất mát
    loss = recon_loss + beta * kl_loss

    return loss, recon_loss, kl_loss

# ===================== PHẦN HUẤN LUYỆN MÔ HÌNH =====================

# def train_cvae_with_params(model, train_loader, val_loader, epochs=100, lr=1e-4, beta=1.0, device=device):
#     """
#     Huấn luyện mô hình cVAE với thông số HR và BR
#     """
#     optimizer = optim.Adam(model.parameters(), lr=lr)
#     scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=5, factor=0.5)

#     # Lưu lịch sử mất mát
#     train_losses = []
#     train_recon_losses = []
#     train_kl_losses = []
#     val_losses = []
#     val_recon_losses = []
#     val_kl_losses = []

#     best_val_loss = float('inf')

#     for epoch in range(epochs):
#         model.train()
#         train_loss = 0
#         train_recon_loss = 0
#         train_kl_loss = 0

#         for batch_idx, (ecg, ppg, resp, hr, rr) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")):
#             ecg = ecg.to(device)
#             ppg = ppg.to(device)
#             resp = resp.to(device)
#             hr = hr.to(device)
#             rr = rr.to(device)

#             # Làm phẳng dữ liệu
#             ecg_flat = ecg.view(ecg.size(0), -1)
#             ppg_flat = ppg.view(ppg.size(0), -1)
#             resp_flat = resp.view(resp.size(0), -1)

#             optimizer.zero_grad()

#             # Forward pass
#             recon_batch, mu, log_var = model(ppg_flat, ecg_flat, resp_flat, hr, rr)

#             # Tính mất mát
#             loss, recon, kl = vae_loss_function(recon_batch, ppg_flat, mu, log_var, beta)

#             # Backward pass
#             loss.backward()
#             optimizer.step()

#             train_loss += loss.item()
#             train_recon_loss += recon.item()
#             train_kl_loss += kl.item()

#         # Đánh giá trên tập validation
#         model.eval()
#         val_loss = 0
#         val_recon_loss = 0
#         val_kl_loss = 0

#         with torch.no_grad():
#             for batch_idx, (ecg, ppg, resp, hr, rr) in enumerate(val_loader):
#                 ecg = ecg.to(device)
#                 ppg = ppg.to(device)
#                 resp = resp.to(device)
#                 hr = hr.to(device)
#                 rr = rr.to(device)

#                 # Làm phẳng dữ liệu
#                 ecg_flat = ecg.view(ecg.size(0), -1)
#                 ppg_flat = ppg.view(ppg.size(0), -1)
#                 resp_flat = resp.view(resp.size(0), -1)

#                 # Forward pass
#                 recon_batch, mu, log_var = model(ppg_flat, ecg_flat, resp_flat, hr, rr)

#                 # Tính mất mát
#                 loss, recon, kl = vae_loss_function(recon_batch, ppg_flat, mu, log_var, beta)

#                 val_loss += loss.item()
#                 val_recon_loss += recon.item()
#                 val_kl_loss += kl.item()

#         # Tính trung bình mất mát
#         train_loss /= len(train_loader.dataset)
#         train_recon_loss /= len(train_loader.dataset)
#         train_kl_loss /= len(train_loader.dataset)

#         val_loss /= len(val_loader.dataset)
#         val_recon_loss /= len(val_loader.dataset)
#         val_kl_loss /= len(val_loader.dataset)

#         # Cập nhật scheduler
#         scheduler.step(val_loss)

#         # Lưu lại lịch sử mất mát
#         train_losses.append(train_loss)
#         train_recon_losses.append(train_recon_loss)
#         train_kl_losses.append(train_kl_loss)
#         val_losses.append(val_loss)
#         val_recon_losses.append(val_recon_loss)
#         val_kl_losses.append(val_kl_loss)

#         # In mất mát
#         print(f"Epoch {epoch+1}/{epochs}")
#         print(f"Train Loss: {train_loss:.4f} (Recon: {train_recon_loss:.4f}, KL: {train_kl_loss:.4f})")
#         print(f"Val Loss: {val_loss:.4f} (Recon: {val_recon_loss:.4f}, KL: {val_kl_loss:.4f})")

#         # Lưu mô hình tốt nhất
#         if val_loss < best_val_loss:
#             best_val_loss = val_loss
#             torch.save(model.state_dict(), os.path.join(MODEL_SAVE_PATH, 'best_cvae_with_params.pth'))
#             print("Đã lưu mô hình tốt nhất!")

#         # Lưu mô hình cuối cùng
#         torch.save(model.state_dict(), os.path.join(MODEL_SAVE_PATH, 'last_cvae_with_params.pth'))

#         # Trực quan hóa kết quả mỗi 10 epoch
#         if (epoch + 1) % 10 == 0 or epoch == 0:
#             visualize_results(model, val_loader, epoch + 1, dataset=val_loader.dataset)

#     # Vẽ đồ thị mất mát
#     plot_loss_curves(train_losses, val_losses,
#                      train_recon_losses, val_recon_losses,
#                      train_kl_losses, val_kl_losses)

#     return model, (train_losses, val_losses)

def vae_loss_function(recon_x, x, mu, log_var, beta=1.0):
    # Thêm epsilon để tránh log(0)
    epsilon = 1e-8

    # Reconstruction loss (MSE) với giới hạn
    recon_loss = F.mse_loss(recon_x, x, reduction='sum')
    recon_loss = torch.clamp(recon_loss, min=-1e4, max=1e4)

    # KL divergence với giới hạn
    kl_loss = -0.5 * torch.sum(1 + log_var - mu.pow(2) - torch.clamp(log_var.exp(), min=epsilon, max=1e4))
    kl_loss = torch.clamp(kl_loss, min=-1e4, max=1e4)

    # Tổng mất mát
    loss = recon_loss + beta * kl_loss

    return loss, recon_loss, kl_loss

def train_cvae_with_params(model, train_loader, val_loader, epochs=100, lr=1e-4, beta=1.0, device=device):
    optimizer = optim.Adam(model.parameters(), lr=lr)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=5, factor=0.5)

    train_losses = []
    val_losses = []

    best_val_loss = float('inf')

    for epoch in range(epochs):
        model.train()
        total_train_loss = 0.0
        total_train_samples = 0

        for batch_idx, (ecg, ppg, resp, hr, rr) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")):
            ecg = ecg.to(device)
            ppg = ppg.to(device)
            resp = resp.to(device)
            hr = hr.to(device)
            rr = rr.to(device)

            # Làm phẳng dữ liệu
            ecg_flat = ecg.view(ecg.size(0), -1)
            ppg_flat = ppg.view(ppg.size(0), -1)
            resp_flat = resp.view(resp.size(0), -1)

            optimizer.zero_grad()

            # Forward pass
            try:
                recon_batch, mu, log_var = model(ppg_flat, ecg_flat, resp_flat, hr, rr)

                # Tính mất mát
                loss, recon, kl = vae_loss_function(recon_batch, ppg_flat, mu, log_var, beta)

                # Kiểm tra NaN
                if torch.isnan(loss):
                    print(f"NaN detected in training loss at epoch {epoch}, batch {batch_idx}")
                    continue

                # Backward pass
                loss.backward()

                # Gradient clipping
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

                optimizer.step()

                total_train_loss += loss.item() * ecg.size(0)
                total_train_samples += ecg.size(0)

            except Exception as e:
                print(f"Error in training: {e}")
                continue

        # Tính trung bình mất mát
        train_loss = total_train_loss / max(total_train_samples, 1)

        # Validation
        model.eval()
        total_val_loss = 0.0
        total_val_samples = 0

        with torch.no_grad():
            for batch_idx, (ecg, ppg, resp, hr, rr) in enumerate(val_loader):
                ecg = ecg.to(device)
                ppg = ppg.to(device)
                resp = resp.to(device)
                hr = hr.to(device)
                rr = rr.to(device)

                # Làm phẳng dữ liệu
                ecg_flat = ecg.view(ecg.size(0), -1)
                ppg_flat = ppg.view(ppg.size(0), -1)
                resp_flat = resp.view(resp.size(0), -1)

                if torch.isnan(ecg).any() or torch.isinf(ecg).any():
                  print(f"NaN or Inf detected in ECG at epoch {epoch}, batch {batch_idx}")
                if torch.isnan(ppg).any() or torch.isinf(ppg).any():
                  print(f"NaN or Inf detected in PPG at epoch {epoch}, batch {batch_idx}")
                if torch.isnan(resp).any() or torch.isinf(resp).any():
                  print(f"NaN or Inf detected in RESP at epoch {epoch}, batch {batch_idx}")
                if torch.isnan(hr).any() or torch.isinf(hr).any():
                  print(f"NaN or Inf detected in HR at epoch {epoch}, batch {batch_idx}")
                if torch.isnan(rr).any() or torch.isinf(rr).any():
                  print(f"NaN or Inf detected in RR at epoch {epoch}, batch {batch_idx}")

                # Forward pass
                try:
                    recon_batch, mu, log_var = model(ppg_flat, ecg_flat, resp_flat, hr, rr)

                    # Tính mất mát
                    loss, recon, kl = vae_loss_function(recon_batch, ppg_flat, mu, log_var, beta)

                    total_val_loss += loss.item() * ecg.size(0)
                    total_val_samples += ecg.size(0)

                except Exception as e:
                    print(f"Error in validation: {e}")
                    continue

        # Tính trung bình mất mát validation
        val_loss = total_val_loss / max(total_val_samples, 1)

        # In thông tin mất mát
        print(f"Epoch {epoch+1}: Train Loss = {train_loss}, Val Loss = {val_loss}")

        # Lưu mất mát
        train_losses.append(train_loss)
        val_losses.append(val_loss)

        # Cập nhật scheduler
        scheduler.step(val_loss)

        # Lưu mô hình tốt nhất
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), os.path.join(MODEL_SAVE_PATH, 'best_cvae_with_params.pth'))

    return model, (train_losses, val_losses)

def plot_loss_curves(train_losses, val_losses, train_recon_losses, val_recon_losses, train_kl_losses, val_kl_losses):
    """
    Vẽ đồ thị mất mát
    """
    plt.figure(figsize=(15, 5))

    # Đồ thị tổng mất mát
    plt.subplot(1, 3, 1)
    plt.plot(train_losses, label='Train')
    plt.plot(val_losses, label='Validation')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Total Loss')
    plt.legend()

    # Đồ thị reconstruction loss
    plt.subplot(1, 3, 2)
    plt.plot(train_recon_losses, label='Train')
    plt.plot(val_recon_losses, label='Validation')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Reconstruction Loss')
    plt.legend()

    # Đồ thị KL loss
    plt.subplot(1, 3, 3)
    plt.plot(train_kl_losses, label='Train')
    plt.plot(val_kl_losses, label='Validation')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('KL Divergence Loss')
    plt.legend()

    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_PATH, 'cvae_loss_curves.png'))
    plt.close()

def plot_ppg_spectrum(original_ppg, reconstructed_ppg, fs=125, title="Phổ tín hiệu PPG trước và sau khi qua mô hình"):
    """
    Vẽ phổ tín hiệu PPG gốc và PPG đã tái tạo bằng mô hình.

    Parameters:
    - original_ppg: Tín hiệu PPG ban đầu (trước khi vào model)
    - reconstructed_ppg: Tín hiệu PPG được tái tạo từ model
    - fs: Tần số lấy mẫu (mặc định 125Hz)
    """
    # Tính FFT
    original_fft = np.abs(np.fft.fft(original_ppg))
    reconstructed_fft = np.abs(np.fft.fft(reconstructed_ppg))

    # Trục tần số
    freqs = np.fft.fftfreq(len(original_ppg), d=1/fs)

    # Lọc chỉ lấy phần tần số dương
    mask = freqs > 0
    freqs = freqs[mask]
    original_fft = original_fft[mask]
    reconstructed_fft = reconstructed_fft[mask]

    # Vẽ biểu đồ phổ tín hiệu
    plt.figure(figsize=(10, 5))
    plt.plot(freqs, original_fft, label="PPG Gốc", alpha=0.7)
    plt.plot(freqs, reconstructed_fft, label="PPG Tái tạo", alpha=0.7, linestyle='dashed')
    plt.xlabel("Tần số (Hz)")
    plt.ylabel("Biên độ")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

def compare_ppg_spectrum(model, val_loader, device):
    """
    So sánh phổ PPG trước và sau khi qua mô hình.
    """
    model.eval()  # Chuyển mô hình sang chế độ đánh giá
    with torch.no_grad():
        for batch_idx, (ecg, ppg, resp, hr, rr) in enumerate(val_loader):
            # Chỉ lấy một batch đầu tiên
            ecg = ecg.to(device)
            ppg = ppg.to(device)
            resp = resp.to(device)
            hr = hr.to(device)
            rr = rr.to(device)

            # Làm phẳng dữ liệu
            ecg_flat = ecg.view(ecg.size(0), -1)
            ppg_flat = ppg.view(ppg.size(0), -1)
            resp_flat = resp.view(resp.size(0), -1)

            # 🛠 In dữ liệu đầu vào
            print(f"Batch {batch_idx}:")
            print(f"  ECG shape: {ecg.shape}, Flattened: {ecg_flat.shape}")
            print(f"  PPG shape: {ppg.shape}, Flattened: {ppg_flat.shape}")
            print(f"  Resp shape: {resp.shape}, Flattened: {resp_flat.shape}")
            print(f"  HR: {hr.cpu().numpy()}, RR: {rr.cpu().numpy()}")

            # Forward qua mô hình
            recon_ppg, _, _ = model(ppg_flat, ecg_flat, resp_flat, hr, rr)

            # Chuyển về numpy để vẽ phổ
            original_ppg = ppg_flat.cpu().numpy()[0]  # Lấy tín hiệu đầu tiên
            reconstructed_ppg = recon_ppg.cpu().numpy()[0]  # Lấy tín hiệu đầu tiên

            # 🛠 In dữ liệu sau khi qua model
            print(f"  Original PPG sample (first 5 values): {original_ppg}")
            print(f"  Reconstructed PPG sample (first 5 values): {reconstructed_ppg}")

            # Vẽ phổ tín hiệu
            plot_ppg_spectrum(original_ppg, reconstructed_ppg)
            break  # Chỉ lấy một batch

def visualize_results(model, data_loader, epoch, num_samples=3, dataset=None):
    """
    Trực quan hóa kết quả của mô hình với các thông số HR và BR khác nhau
    """
    model.eval()

    # Lấy một batch dữ liệu
    ecg, ppg, resp, hr, rr = next(iter(data_loader))

    # Chọn một số mẫu
    ecg = ecg[:num_samples].to(device)
    ppg = ppg[:num_samples].to(device)
    resp = resp[:num_samples].to(device)
    hr_orig = hr[:num_samples].to(device)
    rr_orig = rr[:num_samples].to(device)

    # Làm phẳng dữ liệu
    ecg_flat = ecg.view(ecg.size(0), -1)
    ppg_flat = ppg.view(ppg.size(0), -1)
    resp_flat = resp.view(resp.size(0), -1)

    # Tạo các giá trị HR và BR khác nhau để kiểm tra
    hr_values = [-1.0, 0.0, 1.0]  # Giá trị chuẩn hóa (thấp, trung bình, cao)
    rr_values = [-1.0, 0.0, 1.0]  # Giá trị chuẩn hóa (thấp, trung bình, cao)

    # Giải mã giá trị HR và RR thực tế
    hr_actual = []
    rr_actual = []

    for i in range(num_samples):
        if dataset:
            hr_val, rr_val = dataset.denormalize_params(hr_orig[i].item(), rr_orig[i].item())
            hr_actual.append(hr_val)
            rr_actual.append(rr_val)
        else:
            hr_actual.append(f"Norm: {hr_orig[i].item():.2f}")
            rr_actual.append(f"Norm: {rr_orig[i].item():.2f}")

            # Generate PPG and plot spectrum for different HR and RR values
        col_idx = 2
        for hr_idx, hr_val in enumerate(hr_values):
            for rr_idx, rr_val in enumerate(rr_values):
                # ... (your existing code for generating PPG with different HR and RR) ...

                # Convert to numpy arrays for spectrum plotting
                original_ppg = ppg_flat[i].cpu().detach().numpy()
                generated_ppg = generated[0].cpu().detach().numpy()

                # Plot the spectrum
                plot_spectrum(original_ppg, generated_ppg)

    # Giải mã các giá trị HR và RR được sử dụng trong thử nghiệm
    hr_denorm_values = []
    rr_denorm_values = []

    for hr_val in hr_values:
        for rr_val in rr_values:
            if dataset:
                hr_denorm, rr_denorm = dataset.denormalize_params(hr_val, rr_val)
                hr_denorm_values.append(hr_denorm)
                rr_denorm_values.append(rr_denorm)
            else:
                hr_denorm_values.append(f"Norm: {hr_val:.2f}")
                rr_denorm_values.append(f"Norm: {rr_val:.2f}")

    # Tạo grid để hiển thị
    fig, axs = plt.subplots(num_samples, len(hr_values) * len(rr_values) + 2,
                          figsize=(4 * (len(hr_values) * len(rr_values) + 2), 3 * num_samples))

    for i in range(num_samples):
        # Hiển thị tín hiệu ECG
        axs[i, 0].plot(ecg_flat[i].cpu().numpy())
        axs[i, 0].set_title(f'ECG\nHR={hr_actual[i]}\nRR={rr_actual[i]}')
        axs[i, 0].set_yticks([])

        # Hiển thị tín hiệu PPG thực
        axs[i, 1].plot(ppg_flat[i].cpu().numpy())
        axs[i, 1].set_title('PPG Thực')
        axs[i, 1].set_yticks([])

        # Tạo và hiển thị tín hiệu với các thông số khác nhau
        col_idx = 2
        for hr_idx, hr_val in enumerate(hr_values):
            for rr_idx, rr_val in enumerate(rr_values):
                # Tạo tensor cho HR và RR
                hr_tensor = torch.FloatTensor([[hr_val]]).to(device)
                rr_tensor = torch.FloatTensor([[rr_val]]).to(device)

                # Sinh tín hiệu mới
                generated = model.generate(
                    ecg_flat[i].unsqueeze(0),
                    resp_flat[i].unsqueeze(0),
                    hr_tensor,
                    rr_tensor
                )

                # Hiển thị
                axs[i, col_idx].plot(generated[0].cpu().detach().numpy())
                axs[i, col_idx].set_title(f'PPG Sinh\nHR={hr_denorm_values[col_idx-2]}\nRR={rr_denorm_values[col_idx-2]}')
                axs[i, col_idx].set_yticks([])

                col_idx += 1

    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_PATH, f'cvae_results_epoch_{epoch}.png'))
    plt.close()

def experiment_with_parameter_ranges(model, data_loader, dataset):
    """
    Thí nghiệm với nhiều giá trị HR và BR khác nhau
    """
    # Lấy một mẫu dữ liệu
    ecg, ppg, resp, _, _ = next(iter(data_loader))
    ecg = ecg[0].view(1, -1).to(device)
    resp = resp[0].view(1, -1).to(device)

    # Tạo các giá trị HR và BR để thử nghiệm
    if dataset:
        # Tạo các giá trị thực tế
        hr_values = np.linspace(50, 120, 5)  # Từ 50 đến 120 BPM
        rr_values = np.linspace(8, 30, 5)    # Từ 8 đến 30 BrPM

        # Chuẩn hóa các giá trị
        hr_norm = [(hr - dataset.hr_mean) / dataset.hr_std for hr in hr_values]
        rr_norm = [(rr - dataset.rr_mean) / dataset.rr_std for rr in rr_values]
    else:
        # Sử dụng trực tiếp giá trị chuẩn hóa
        hr_norm = np.linspace(-2.0, 2.0, 5)
        rr_norm = np.linspace(-2.0, 2.0, 5)
        hr_values = hr_norm
        rr_values = rr_norm

    # Tạo grid để hiển thị
    fig, axs = plt.subplots(len(hr_norm), len(rr_norm), figsize=(15, 15))

    for i, hr in enumerate(hr_norm):
        for j, rr in enumerate(rr_norm):
            # Sinh tín hiệu
            hr_tensor = torch.FloatTensor([[hr]]).to(device)
            rr_tensor = torch.FloatTensor([[rr]]).to(device)

            generated = model.generate(ecg, resp, hr_tensor, rr_tensor)

            # Hiển thị
            axs[i, j].plot(generated[0].cpu().detach().numpy())
            if dataset:
                axs[i, j].set_title(f'HR={hr_values[i]:.1f} BPM, RR={rr_values[j]:.1f} BrPM')
            else:
                axs[i, j].set_title(f'HR norm={hr:.2f}, RR norm={rr:.2f}')
            axs[i, j].set_xticks([])
            axs[i, j].set_yticks([])

    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_PATH, 'parameter_range_experiment.png'))
    plt.close()

# ===================== PHẦN CHẠY THỬ NGHIỆM =====================

def run_experiment(epochs=30):
    """
    Chạy thử nghiệm mô hình cVAE với thông số HR và BR
    """
    # Thông số
    batch_size = 32
    test_size = 0.2
    input_dim = 1250  # Kích thước cửa sổ 10s với tần số lấy mẫu 125Hz
    latent_dim = 128
    hidden_dim = 512

    # Chuẩn bị dữ liệu
    train_loader, val_loader, dataset = prepare_dataloaders(batch_size, test_size)

    # Khởi tạo mô hình
    print("\n=== Khởi tạo mô hình Conditional VAE với thông số HR và BR ===")
    cvae_params = ConditionalVAEWithParams(input_dim, hidden_dim, latent_dim, input_dim).to(device)

    # Huấn luyện mô hình
    print("\n=== Bắt đầu huấn luyện ===")
    model, losses = train_cvae_with_params(cvae_params, train_loader, val_loader, epochs=epochs, lr=1e-4, beta=0.5, device=device)
    compare_ppg_spectrum(model, val_loader, device)

    # Thử nghiệm với các thông số khác nhau
    print("\n=== Thí nghiệm với các thông số HR và BR khác nhau ===")
    experiment_with_parameter_ranges(model, val_loader, dataset)

    print("\n=== Hoàn thành ===")
    return model, dataset

if __name__ == "__main__":
    # Xử lý dữ liệu nếu chưa có
    if not os.path.exists(os.path.join(PROCESSED_PATH, 'bidmc_processed_segments.npy')):
        print("Bắt đầu xử lý dữ liệu BIDMC...")
        process_bidmc_data()

    # Chạy thử nghiệm
    print("Bắt đầu huấn luyện mô hình cVAE...")
    model, dataset = run_experiment(epochs=30)
```
